task: Write a short Python script that opens the file, tokenizes the text, and filters out all non-alphabetic tokens (punctuation, numbers). All other tokens (i.e. only the plain words) shall be retained and saved in an output file. 
Upload your script (.ipynb or .py) as solution.

```python
#first step: loading the text
with open("eu_referendum-1.txt","r") as f:
    txt = f.read()
    print(txt)
```

```python
#second step: tokenization of text
from nltk.tokenize import word_tokenize
tokens = word_tokenize(txt)
print(tokens)
```

```python
#third step: filtering the non-alphabetic tokens out of te text

import string, re 
s = txt
table = str.maketrans({key: None for key in string.punctuation})
new_s = s.translate(table)  
#print(new_s)

plaintxt_1 = re.sub(r'[^\w\s]','',new_s)
plaintxt_2 = re.sub(r'[^\D]','',plaintxt_1)
new_tokens = word_tokenize(plaintxt_2)
print(new_tokens)

## first try, i left it in for my orientation
#import regex
#s = txt
#remove = regex.compile(ur'[\p{C}|\p{M}|\p{P}|\p{S}|\p{Z}]+', regex.UNICODE)
#remove.sub(u" ", s).strip()
```

```python
#fourth step: preparing the output file
with open("result_eLinek.txt","w") as f:
    f.write("\n".join(new_tokens))   
    
```
