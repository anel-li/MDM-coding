```python
library(shiny)
library(leaflet)
library(dplyr)
```

```python
server <- function(input, output, session){
    
  data <- reactive({
    x <- manhattan
  })
  
  output$mymap <- renderLeaflet({
    df <- data()
    
    m <- leaflet(data = df) %>%
      addTiles() %>%
      addMarkers(
                 clusterOptions = markerClusterOptions(),
                 lng = ~Longitude,
                 lat = ~Latitude,
                 popup = paste("Location:", manhattan$place_full_name, "<br>",
                               "Time Stamp:", manhattan$created_at, "<br>",
                               "Tweet:", manhattan$text))
    m
  })
}
```

```python
load("manhattan.Rda")
```

We split the geo information into two variables:

```python
manhattan$geo_coords
```

```python
manhattan <- tidyr::separate(
    data = manhattan,
    col = geo_coords,
    into = c("Latitude", "Longitude"),
    sep = ",",
    remove = FALSE)
```

```python
manhattan$Latitude
```

We remove the bracket from the geo coordinates.

```python
manhattan$Latitude <- stringr::str_replace_all(manhattan$Latitude, "[c(]", "")
```

```python
manhattan$Longitude <- stringr::str_replace_all(manhattan$Longitude, "[)]", "")
```

And convert the information from string to numeric.

```python
manhattan$Latitude <- as.numeric(manhattan$Latitude)
```

```python
manhattan$Longitude <- as.numeric(manhattan$Longitude)
```

Next we remove some posts outside of our (very specific) bounding box.

```python
manhattan<-manhattan[(manhattan$Latitude >= 40.698136 & manhattan$Latitude <= 40.798978),]
```

```python
manhattan<-manhattan[(manhattan$Longitude >= -74.022208 & manhattan$Longitude <= -73.936377),]
```

And save our processed data.

```python
save(manhattan,file = "manhattan_geo.Rda")
```

```python
ui <- leafletOutput("mymap", height = 1000)
```

```python
shinyApp(ui = ui, server = server)
```
