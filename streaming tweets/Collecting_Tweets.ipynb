Let's do the same, but this time we store the raw data (we do not parse it) in the file `mytest.json`.

```python
stream_tweets(
    "",
    timeout = 5,
    file_name = "sample.json",
    parse = FALSE)
```

We can take a look how the raw data looks by checking the file...

Now let's parse it back into R using the `stream_parse` function from rtweets.

```python
sample <- parse_stream("sample.json")
sample
```

Now we look for some keyswords that we are intersted in.

```python
stream_tweets(
    q = "vaccine, vaccination",
    timeout = 10,
    file_name = "vaccine.json",
    parse = FALSE
)
```

Again we can parse our data from the file.

```python
vaccine <- parse_stream("vaccine.json")
```

To keep the data after modifying it we can save it as a R data file as well (.Rda).

```python
save(vaccine, file = "vaccine_streamed.Rda")
```

After removing it...

```python
rm(vaccine)
```

We can bring it back easily using `load`.

```python
load("vaccine_streamed.Rda")
head(vaccine)
```

Lastly we collect some tweets using a bounding box. For that we need the geo-coordinates first. We can get those for example from

https://boundingbox.klokantech.com/ we select `CSV` to receive the coordinates in a simple format.

Instead of specifiying keywords, we can specify a bounding-box.

```python
stream_tweets(
    c(5.9559,45.818,10.4921,47.8084),
    timeout = 60,
    file_name = "ulm.json",
    parse = FALSE
)
```

```python
ulm <- parse_stream("ulm.json")
```
