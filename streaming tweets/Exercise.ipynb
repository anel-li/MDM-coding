Collect tweets for 120 seconds using one or multiple keywords and store it in a file called `mytweets.json`.

```python
stream_tweets(
    "",
    timeout = 120,
    file_name = "ex.json",
    parse = FALSE)
```

Use `parse_stream` to load the tweets into an object `mytweets`.

```python
ex <- parse_stream("ex.json")
ex

save(ex, file = "ex.Rda")
```

Remove links from the Tweets.

```python
library(tidytext)
library(dplyr)

load("ex.Rda")
nrow(ex)

ex$text <- gsub("http.*", "", ex$text)
ex$text <- gsub("https.*", "", ex$text)
ex$text <- gsub("&amp;", "&", ex$text)
```

Use `unnest_tokens` to get a list of all the words.

```python
ex_clean <- ex %>%
  dplyr::select(text) %>%
  unnest_tokens(word, text)

head(ex_clean)
```

Load a stopword list. If you want to analyze German tweets you can use a German stopword list.
